```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the webpage
url = "https://www.biopharmadive.com/news/biotech-ipo-performance-tracker/587604/"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table element
table = soup.find('div', class_='ctable__container')

# Extract table headers
headers = [
    "Company",
    "IPO Date",
    "Total Proceeds",
    "IPO Price",
    "Current Price",
    "Change in Price",
    "Current Market Cap"
]

# Extract table rows
rows = []
for row in table.find_all('div', class_='ctable_row'):
    cells = row.find_all('div', class_='table_cell')
    row_data = []
    for cell in cells:
        if 'table_cell--companyname' in cell.get('class', []):
            row_data.append(cell.text.strip())
        elif 'table_cell--ipodate' in cell.get('class', []):
            row_data.append(cell.text.strip())
        elif 'table_cell--totalproceeds' in cell.get('class', []):
            row_data.append(cell.text.strip())
        elif 'table_cell--ipoprice' in cell.get('class', []):
            row_data.append(cell.text.strip())
        elif 'table_cell--currentstockprice' in cell.get('class', []):
            row_data.append(cell.text.strip())
        elif 'table_cell--change' in cell.get('class', []):
            row_data.append(cell.text.strip())
        elif 'table_cell--currentmarketcap' in cell.get('class', []):
            row_data.append(cell.text.strip())
    if row_data:
        rows.append(row_data)

# Create a pandas DataFrame
df = pd.DataFrame(rows, columns=headers)

# Display the DataFrame
print(df)

# Optionally, save to a CSV file
df.to_csv('biotech_ipo_data.csv', index=False)
```
