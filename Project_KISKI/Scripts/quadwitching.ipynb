This code analyzes the stock performance of three major market indices—S&P 500 (SPY), NASDAQ Composite, and Russell 2000—on specific days known as "quad witching days." These are the days when four types of financial contracts expire, which often leads to higher trading volume and volatility.

```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt

# Define the date range
start_date = '1990-01-01'
end_date = datetime.today().strftime('%Y-%m-%d')

# Define the tickers
tickers = {
    "SPY": "SPY",
    "NASDAQ": "^IXIC",  # NASDAQ Composite
    "Russell": "^RUT",   # Russell 2000
    "NBI": "NBI",   # NBI
    "XBI": "XBI"   # XBI 
}

# Function to fetch data
def fetch_data(ticker, start, end):
    # Use 'Close' instead of 'Adj Close' to exclude dividend adjustments
    return yf.download(ticker, start=start, end=end)['Close']

# Function to determine quad witching days within a given date range
def get_quadwitching_days(start_date, end_date):
    quadwitching_days = []
    start_year = pd.Timestamp(start_date).year
    end_year = pd.Timestamp(end_date).year
    
    for year in range(start_year, end_year + 1):
        for month in [3, 6, 9, 12]:
            third_friday = pd.date_range(start=f'{year}-{month}-01', 
                                         end=f'{year}-{month}-30', 
                                         freq='WOM-3FRI')
            if not third_friday.empty:
                quadwitching_days.append(third_friday[0])
    
    return quadwitching_days

# Fetch quad witching days
quadwitching_days = get_quadwitching_days(start_date, end_date)

# Fetch data for the defined tickers
data = {key: fetch_data(ticker, start_date, end_date) for key, ticker in tickers.items()}
df = pd.DataFrame(data)

# Filter for quad witching days
quad_df = df[df.index.isin(quadwitching_days)]

# Calculate daily returns
returns = quad_df.pct_change().dropna()

# Function to calculate Sharpe Ratio
def calculate_sharpe_ratio(returns, risk_free_rate=0.0):
    excess_returns = returns - risk_free_rate
    return excess_returns.mean() / excess_returns.std()

# Function to calculate maximum drawdown
def calculate_max_drawdown(returns):
    cumulative_returns = (1 + returns).cumprod()
    peak = cumulative_returns.cummax()
    drawdown = (cumulative_returns - peak) / peak
    return drawdown.min()

# Calculate performance metrics for quad witching days
performance = {}
for index in returns.columns:
    up_days = (returns[index] > 0).sum()
    down_days = (returns[index] < 0).sum()
    total_days = len(returns[index])
    avg_return = returns[index].mean()
    std_dev = returns[index].std()
    max_return = returns[index].max()
    min_return = returns[index].min()
    
    sharpe_ratio = calculate_sharpe_ratio(returns[index])
    max_drawdown = calculate_max_drawdown(returns[index])

    performance[index] = {
        "Up Days %": (up_days / total_days) * 100,
        "Down Days %": (down_days / total_days) * 100,
        "Average Return": avg_return,
        "Standard Deviation": std_dev,
        "Max Return": max_return,
        "Min Return": min_return,
        "Total Days": total_days,
        "Sharpe Ratio": sharpe_ratio,
        "Max Drawdown": max_drawdown
    }

# Display results with enhanced information
for index, stats in performance.items():
    print(f"{index}:")
    print(f"  Up Days %: {stats['Up Days %']:.2f}%")
    print(f"  Down Days %: {stats['Down Days %']:.2f}%")
    print(f"  Average Return: {stats['Average Return']:.4f}")
    print(f"  Standard Deviation: {stats['Standard Deviation']:.4f}")
    print(f"  Max Return: {stats['Max Return']:.4f}")
    print(f"  Min Return: {stats['Min Return']:.4f}")
    #print(f"  Sharpe Ratio: {stats['Sharpe Ratio']:.4f}")
    print(f"  Max Drawdown: {stats['Max Drawdown']:.4f}")
    print(f"  Total Quad Witching Days: {stats['Total Days']}\n")

# Plot performance comparison for quad witching days
performance_df = pd.DataFrame(performance).T

plt.figure(figsize=(10, 5))

performance_df['Average Return'].plot(kind='bar', color='skyblue', alpha=0.7)
plt.title('Average Return on Quad Witching Days')
plt.xlabel('Ticker')
plt.ylabel('Average Return')
plt.axhline(0, color='red', linewidth=0.8, linestyle='--')

plt.tight_layout()
plt.show()
```
