```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import backtrader as bt
from backtrader.feeds import PandasData
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, KFold
from tscv import GapKFold
import pyfolio as pf
%matplotlib inline

warnings.filterwarnings("ignore")
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 150)
pd.options.display.float_format = "{:.2f}".format
plt.style.use('ggplot')
sns.set(style='darkgrid', context='talk', palette='Dark2')

def fetch_data():
    dataset = yf.download('^NBI', start='1994-01-01')
    dataset = dataset.sort_index(ascending=True)
    return dataset

def prepare_data(dataset, quantile=0.5):
    df = dataset.copy()
    df['closeGap'] = df['Close'].pct_change()
    df['highGap'] = df['High'].pct_change()
    df['lowGap'] = df['Low'].pct_change()
    df['volGap'] = df['Volume'].pct_change()
    df['dailyChange'] = (df['Close'] - df['Open']) / df['Open']
    df['nextDayDirection'] = (df['Volume'].shift(-1) - df['Volume'])
    df = df.dropna()
    
    percentile_next_day_direction = df['nextDayDirection'].quantile(quantile)
    
    features = ['closeGap', 'highGap', 'lowGap', 'volGap', 'dailyChange']
    X = df[features]
    y = pd.Series(np.where((df.nextDayDirection > percentile_next_day_direction), 1, 0), index=df.index)
    
    print(f"\n{quantile*100}th percentile value used for classification: {percentile_next_day_direction:.2f}")
    
    return X, y, df

def plot_signals(df, year='2023', save_path=None):
    df = df[f'{year}-01-01':]
    buys = df.loc[df["nextDayDirection"] > 0]
    sells = df.loc[df["nextDayDirection"] <= 0]

    plt.figure(figsize=(20, 5))
    plt.plot(df['Close'], lw=2., label='Price')
    plt.plot(buys.index, buys['Close'], '^', markersize=10, color='red', lw=2., label='Buy')
    plt.plot(sells.index, sells['Close'], 'v', markersize=10, color='green', lw=2., label='Sell')
    plt.ylabel('Price (USD)')
    plt.xlabel('Date')
    plt.title(f'Buy and Sell Signals ({year})')
    plt.legend(loc='best')
    plt.grid(True)
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.close()
    else:
        plt.show()

def clean_data(x_data, y_data):
    if isinstance(x_data, pd.DataFrame) or isinstance(x_data, pd.Series):
        original_index = x_data.index
        x_data = x_data.replace([np.inf, -np.inf], np.nan).dropna()
        if y_data is not None:
            if isinstance(y_data, np.ndarray):
                y_data = pd.Series(y_data, index=original_index)
            y_data = y_data[x_data.index]
    elif isinstance(x_data, np.ndarray):
        mask = np.isfinite(x_data).all(axis=1)
        x_data = x_data[mask]
        if y_data is not None:
            y_data = y_data[mask]
    return x_data, y_data

def train_model(X, y, model, params=None):
    gkcv = GapKFold(n_splits=3, gap_before=2, gap_after=1)
    for train_samples, test_samples in gkcv.split(X):
        X_train, X_test = X.iloc[train_samples], X.iloc[test_samples]
        y_train, y_test = y[train_samples], y[test_samples]

    X_train_clean, y_train_clean = clean_data(X_train, y_train)
    X_test_clean, _ = clean_data(X_test, None)

    if model == 'decision_tree':
        clf = DecisionTreeClassifier(**(params or {}))
    elif model == 'random_forest':
        clf = RandomForestClassifier(**(params or {}))
    elif model == 'svc':
        clf = SVC(**(params or {}))
    elif model == 'gradient_boosting':
        clf = GradientBoostingClassifier(**(params or {}))
    else:
        raise ValueError("Unsupported model type.")

    clf.fit(X_train_clean, y_train_clean)
    cv_results = cross_val_score(clf, X_train_clean, y_train_clean, cv=KFold(n_splits=5), scoring='accuracy')
    print(f"Model {model} Accuracy: {cv_results.mean():.4f} (+/- {cv_results.std() * 2:.4f})")
    
    return clf, X_test_clean

def generate_predictions(model, X_test_clean, dataset):
    predictions = pd.DataFrame(model.predict(X_test_clean), columns=['predictedSignal'], index=X_test_clean.index)
    z = dataset.loc[X_test_clean.index, ['Open', 'High', 'Low', 'Close', 'Volume']]
    prices_df = pd.concat([z, predictions], axis=1)
    prices_df['predictedSignal'] = prices_df['predictedSignal'].diff().fillna(0)
    
    return prices_df

def plot_predictions(prices_df, year='2023', save_path=None):
    dd = prices_df[f'{year}-01-01':]
    sells = dd.loc[dd['predictedSignal'] == -1]
    buys = dd.loc[dd['predictedSignal'] == 1]

    plt.figure(figsize=(20, 5))
    plt.plot(dd.index, dd['Close'], lw=.5, label='Price')
    plt.plot(buys.index, dd.loc[buys.index]['Close'], 'v', markersize=10, color='red', lw=2., label='Buy')
    plt.plot(sells.index, dd.loc[sells.index]['Close'], '^', markersize=10, color='green', lw=2., label='Sell')
    plt.ylabel('Price (USD)')
    plt.xlabel('Date')
    plt.title(f'Buy and Sell signals (year {year})')
    plt.legend(loc='best')
    plt.grid(True)
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.close()
    else:
        plt.show()

class SignalData(PandasData):
    cols = ['open', 'high', 'low', 'close', 'volume', 'predictedSignal']
    lines = tuple(cols)
    params = {c: -1 for c in cols}
    params.update({'Date': None})
    params = tuple(params.items())

class TradingStrategy(bt.Strategy):
    params = (
        ('sma_period', 20),  # Example parameter for a moving average strategy
        ('stop_loss', 0.02),  # Example parameter for a stop loss percentage
        ('take_profit', 0.05) # Example parameter for a take profit percentage
    )

    def __init__(self):
        self.data_predicted = self.datas[0].predictedSignal
        self.data_open = self.datas[0].open
        self.data_close = self.datas[0].close
        self.data_volume = self.datas[0].volume

        # Example indicators
        self.sma = bt.indicators.SimpleMovingAverage(self.data_close, period=self.params.sma_period)

    def next_open(self):
        if not self.position:
            if self.data_predicted[0] == 1 and self.data_close[0] > self.sma[0]:
                size = int(self.broker.getcash() / self.datas[0].open)
                self.buy(size=size)
        else:
            if self.data_predicted[0] == -1 or self.data_close[0] < self.sma[0]:
                self.sell(size=self.position.size)

def run_backtest(prices_df, strategy_params):
    cerebro = bt.Cerebro(stdstats=False, cheat_on_open=True)
    cerebro.addstrategy(TradingStrategy, **strategy_params)
    cerebro.adddata(SignalData(dataname=prices_df))
    cerebro.broker.setcash(1000000.0)
    cerebro.broker.setcommission(commission=0.001)
    cerebro.addanalyzer(bt.analyzers.PyFolio, _name='pyfolio')

    print(f'Starting Portfolio Value: {cerebro.broker.getvalue():.2f}')
    results = cerebro.run()
    final_value = cerebro.broker.getvalue()
    print(f'Ending Portfolio Value: {final_value:.2f}')
    
    return final_value, results[0]

def analyze_performance(results, dataset):
    strat = results.analyzers.getbyname('pyfolio')
    returns, _, _, _ = strat.get_pf_items()
    returns.name = 'Strategy'

    benchmark_rets = dataset['Adj Close'].pct_change().dropna()
    benchmark_rets.index = benchmark_rets.index.tz_localize('UTC')
    benchmark_rets = benchmark_rets.filter(returns.index)
    benchmark_rets.name = 'Nasdaq Biotechnology Benchmark'

    print("Strategy Performance:")
    pf.show_perf_stats(returns)
    print("\nBenchmark Performance:")
    pf.show_perf_stats(benchmark_rets)

    return returns, benchmark_rets

def run_parameter_grid_search(X, y, dataset):
    models = {
        'decision_tree': {'max_depth': [None, 5, 10, 15]},
        'random_forest': {'n_estimators': [50, 100, 200]},
        'svc': {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']},
        'gradient_boosting': {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]}
    }
    
    strategy_params_list = [
        {'sma_period': 10, 'stop_loss': 0.02, 'take_profit': 0.05},
        {'sma_period': 20, 'stop_loss': 0.03, 'take_profit': 0.07},
        {'sma_period': 30, 'stop_loss': 0.01, 'take_profit': 0.10}
    ]
    
    results = []

    for model_name, params_grid in models.items():
        params_combinations = pd.MultiIndex.from_product(params_grid.values(), names=params_grid.keys())
        for params in params_combinations:
            param_dict = dict(zip(params_grid.keys(), params))
            print(f"Testing {model_name} with params={param_dict}")
            try:
                model, X_test_clean = train_model(X, y, model_name, params=param_dict)
                prices_df = generate_predictions(model, X_test_clean, dataset)
                plot_predictions(prices_df)  # No saving at this stage
                for strategy_params in strategy_params_list:
                    final_value, result = run_backtest(prices_df, strategy_params)
                    returns, benchmark_rets = analyze_performance(result, dataset)
                    
                    results.append({
                        'model': model_name,
                        'params': param_dict,
                        'strategy_params': strategy_params,
                        'final_value': final_value,
                        'returns': returns,
                        'benchmark_rets': benchmark_rets
                    })
            except Exception as e:
                print(f"Error testing {model_name} with params={param_dict}: {e}")

    results_df = pd.DataFrame(results)

    # Find the best strategy based on final portfolio value
    best_strategies = results_df.nlargest(3, 'final_value')
    worst_strategies = results_df.nsmallest(3, 'final_value')
    
    print("\nTop 3 Strategies:")
    print(best_strategies)

    print("\nBottom 3 Strategies:")
    print(worst_strategies)

    return results_df, best_strategies, worst_strategies

def main():
    dataset = fetch_data()
    quantiles = [0.5]  # You can add more quantiles to test different thresholds
    results_all = []
    best_strategies_all = []
    worst_strategies_all = []

    for quantile in quantiles:
        print(f"\nTesting with quantile: {quantile}")
        X, y, df = prepare_data(dataset, quantile)
        
        # Define training and test periods
        train_start = '1994-01-01'
        train_end = '2021-08-01'
        test_start = '2021-09-02'
        test_end = '2022-08-02'
        backtest_start = '2022-08-03'
        backtest_end = '2024-02-01'

        df_train = df[train_start:train_end]
        df_test = df[test_start:test_end]
        df_backtest = df[backtest_start:backtest_end]

        print(f"Training period: {train_start} to {train_end}")
        print(f"Test period: {test_start} to {test_end}")
        print(f"Backtest period: {backtest_start} to {backtest_end}")

        plot_signals(df, year='2023', save_path=f'original_price_chart_{quantile}.png')
        results_df, best_strategies, worst_strategies = run_parameter_grid_search(X, y, dataset)
        
        # Save images only for the best strategy
        best_strategy = best_strategies.iloc[0]
        best_model = best_strategy['model']
        best_params = best_strategy['params']
        best_strategy_params = best_strategy['strategy_params']
        
        print(f"\nBest Model: {best_model}")
        print(f"Best Model Parameters: {best_params}")
        print(f"Best Strategy Parameters: {best_strategy_params}")

        # Plot and save for the best strategy
        model, X_test_clean = train_model(X, y, best_model, params=best_params)
        prices_df = generate_predictions(model, X_test_clean, dataset)
        plot_predictions(prices_df, save_path=f'best_strategy_predictions_{quantile}.png')

        # Analyze performance for the best strategy
        final_value, result = run_backtest(prices_df, best_strategy_params)
        returns, benchmark_rets = analyze_performance(result, dataset)
        
        # Plot performance metrics
        fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(16, 9), constrained_layout=True)
        axes = ax.flatten()

        pf.plot_rolling_beta(returns=returns, factor_returns=benchmark_rets, ax=axes[0])
        axes[0].set_title('Rolling Beta')

        pf.plot_rolling_volatility(returns=returns, factor_returns=benchmark_rets, ax=axes[1])
        axes[1].set_title('Rolling Volatility')

        pf.plot_annual_returns(returns=returns, ax=axes[2])
        axes[2].set_title('Annual Returns')

        pf.plot_monthly_returns_heatmap(returns=returns, ax=axes[3])
        axes[3].set_title('Monthly Returns Heatmap')

        plt.savefig(f'best_strategy_performance_{quantile}.png', dpi=300, bbox_inches='tight')
        plt.close()

        # Save the top and bottom strategies to CSV
        best_strategies.to_csv(f'top_strategies_{quantile}.csv', index=False)
        worst_strategies.to_csv(f'bottom_strategies_{quantile}.csv', index=False)

        results_all.append(results_df)
        best_strategies_all.append(best_strategies)
        worst_strategies_all.append(worst_strategies)

    results_all_df = pd.concat(results_all, ignore_index=True)
    best_strategies_all_df = pd.concat(best_strategies_all, ignore_index=True)
    worst_strategies_all_df = pd.concat(worst_strategies_all, ignore_index=True)

    # Save combined results
    results_all_df.to_csv('all_results.csv', index=False)
    best_strategies_all_df.to_csv('all_top_strategies.csv', index=False)
    worst_strategies_all_df.to_csv('all_bottom_strategies.csv', index=False)

if __name__ == "__main__":
    main()
 
```
